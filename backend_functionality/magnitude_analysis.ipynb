```python
import numpy as np
import scipy.io.wavfile as wav
import librosa as lb
import matplotlib.pyplot as plt 
```

```python
raw_audio, sample_rate = lb.load('sample1.wav')
```

```python
print(raw_audio)
```

```python
print(sample_rate)
```

```python
lb.get_samplerate('sample1.wav')
```

```python
audio_stft = lb.stft(raw_audio)
```

```python
magnitude, phase = lb.magphase(audio_stft) # maybe can ignore phase because not really using it?
magnitude
```

```python
audio_db = lb.amplitude_to_db(magnitude)
audio_db
```

```python
plt.figure(figsize=(10, 6))
plt.imshow(np.abs(audio_db), aspect='auto', origin='lower', cmap='viridis')
plt.colorbar(format='%+2.0f dB')
plt.title('Magnitude Spectrum of STFT')
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
